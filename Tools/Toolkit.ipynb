```python
from langchain_community.tools import tool
```

```python
@tool
def multiply(a:int,b:int) -> int:
    """Multiply two numbers"""
    return a*b

@tool
def add(a:int,b:int) -> int:
    """Multiply two numbers"""
    return a+b
```

```python
class MathToolkit:
    def get_tools(self):
        return [add,multiply]
```

```python
toolkit = MathToolkit()
tools = toolkit.get_tools()

for tool in tools:
    print(tool.name)
```
