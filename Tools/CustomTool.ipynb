```python
from langchain_ollama import ChatOllama
from langchain_community.tools import tool
```

```python
@tool
def multiply(a:int,b:int) -> int:
    """Multiply two numbers"""
    return a*b
```

```python
print(multiply.invoke({'a':5,'b':15}))
```

```python
from langchain_community.tools import StructuredTool
from pydantic import BaseModel,Field
```

```python
class Multiplyinputs(BaseModel):
    a:int =Field(required=True,description='First NUmber to add')
    b:int =Field(required=True,description='Second NUmber to add')
```

```python
def multiply_function(a,b):
    return a*b
```

```python
multiply_tool = StructuredTool.from_function(
    func=multiply_function,
    name='multiply',
    description='multiply two numbers',
    args_schema=Multiplyinputs
)
```

```python
result=multiply_tool.invoke({'a':5,'b':5})

print(result)
```

```python
from langchain_community.tools import BaseTool
from typing import Type
```

```python
class Multiplyinputs(BaseModel):
    a:int =Field(required=True,description='First NUmber to add')
    b:int =Field(required=True,description='Second NUmber to add')
```

```python
class MultiplyTool(BaseTool):
    name:str = 'multiply'
    description:str = 'multiply two numbers'
    args_schema:Type[BaseModel]=Multiplyinputs
    def _run(self,a:int,b:int)->int:
        return a*b
```

```python
multiply_tool = MultiplyTool()
```

```python
result = multiply_tool.invoke({'a':5,'b':10})
```

```python
print(result)
```
